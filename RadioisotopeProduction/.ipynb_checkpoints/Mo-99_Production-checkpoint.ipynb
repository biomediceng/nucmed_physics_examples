```python
import matplotlib.pyplot as plt
from numpy import exp,log,arange

%matplotlib inline
```

```python
def decay_const(hl):
    return log(2)/hl
```

# Define constants for production and decay

```python
R = 1467.942 #Nuclei/day
hl_isotope=2.75 #days
lambda_isotope = decay_const(hl_isotope)
```

```python
t=arange(0,8)
t2=arange(7,16)
t3=arange(0,16)
t4=arange(0,30)
```

```python
def production(prod_rate,lambda_isotope,t):
    return prod_rate*(1-exp(-lambda_isotope*t))
```

```python
def decay(A_initial,lambda_isotope,t1,t2):
    return A_initial*exp(-lambda_isotope*(t2-t1))
```

```python
A_theoretical=production(R,lambda_isotope,t3)
A_prod=production(R,lambda_isotope,t)
A_decay=decay(A_prod[-1],lambda_isotope,t[-1],t2)
```

```python
plt.plot(t3,A_theoretical,'--b')
plt.plot(t,A_prod,'b')
plt.plot(t2,A_decay,'g')
plt.legend(['$^{99}$Mo Production','$^{99}$Mo Decay'])
# plt.legend(['$^{99}$Mo','$^{99m}$Tc'])
plt.title('$^{99}$Mo Production');
plt.xlabel('t [days]')
plt.ylabel('Mo-99 Activity [GBq]')
plt.xlim([0, t3[-1]])
plt.ylim([0, A_theoretical.max()])
```

```python
A_theoretical=production(R,lambda_isotope,t4)
plt.plot(t4,A_theoretical,'b')
plt.vlines(hl_isotope,0,A_theoretical[t4[(t4>hl_isotope)][0]],linestyle='dashed')
plt.vlines(2*hl_isotope,0,A_theoretical[t4[(t4>2*hl_isotope)][0]],linestyle='dashed')
plt.vlines(3*hl_isotope,0,A_theoretical[t4[(t4>3*hl_isotope)][0]],linestyle='dashed')
plt.vlines(4*hl_isotope,0,A_theoretical[t4[(t4>4*hl_isotope)][0]],linestyle='dashed')
plt.xlim([0, t4[-1]])
plt.ylim([0, A_theoretical.max()])

plt.title('$^{99}$Mo Production');
plt.xlabel('t [days]')
plt.ylabel('Mo-99 Activity [GBq]')
```
