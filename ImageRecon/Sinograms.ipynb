```python
from skimage.transform import radon
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.zeros((91,91))
# a[45,45]=200
```

```python
a[22,22]=200
a[22,78]=200
a[57,35:56]=200
```

```python
a
```

```python
def proj_show(image,sinogram):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4.5))

    ax1.set_title("Original")
    ax1.imshow(image, cmap=plt.cm.Greys_r)

    ax2.set_title("Radon transform\n(Sinogram)")
    ax2.set_xlabel("Projection angle (deg)")
    ax2.set_ylabel("Projection position (pixels)")
    ax2.imshow(sinogram, cmap=plt.cm.Greys_r,
           extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')
    
    ax3.plot(np.arange(sinogram.shape[0]),sinogram[:,0])

    fig.tight_layout()
    plt.show()
```

```python
plt.imshow(a)
```

```python
sino=radon(a,theta=np.linspace(0,3.14,1000),circle=True)
```

```python
proj_show(a,sino)
```

```python
from numpy import cos,sin
```

```python
def xr(x0,y0,phi):
    return x0*cos(phi)+y0*sin(phi)
```

```python
phis=np.linspace(0,3.14,100)
new_x1=xr(-1,1,phis)
new_x2=xr(1,1,phis)
new_x3=xr(-0.5,-0.5,phis)
new_x4=xr(0.5,-0.5,phis)
```

```python
plt.plot(new_x1,phis,'k',new_x2,phis,'r',new_x3,phis,'b',new_x4,phis,'b')
plt.fill_betweenx(phis,new_x3,new_x4,color='blue',alpha=0.3)
plt.legend(['Left Eye','Right Eye','Left Point Mouth','Right Point Mouth'],loc='right',bbox_to_anchor=(1.4,0.5));
plt.title('Sinogram of Face')
plt.xlabel('xr')
plt.ylabel('Angle in Radians')
plt.savefig('Sinogram.png',dpi=300)
```
