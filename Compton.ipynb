```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define the function for Compton Scattering

```python
def compton(E_init,theta):
    new_E = E_init/(1+E_init/511*(1-np.cos(theta)))
    return new_E
```

## And let's try 100 angles between 0 and $\pi$

```python
angles=np.linspace(0,3.14,100)
```

#  How's the energy of the scattered photon as a function of scattering angle for photons of $^{99m}$Tc and $^{131}$I?

```python
Tc99m = compton(140,angles)
I131 = compton(364,angles)
```

```python
plt.plot(angles,Tc99m,angles,I131,xunits='radians')
plt.legend(['Tc-99m','I-131'])
plt.xlim(xmin=0)
plt.grid()
plt.xlabel('Scatter Angle from 0 to $\pi$');
plt.ylabel('Energy of Scattered Photon [keV]')
plt.title('Energy as a function of Compton scatter angle');
```
